```python
num = float(input("Enter a number: "))
sign = "positive" if num > 0 else ("negative" if num < 0 else "zero")
if num == int(num) and num != 0:
    check = "even" if int(num) % 2 == 0 else "odd"
    print(f"{num} is {sign} and {check}")
else:
    print(f"{num} is {sign}")
```

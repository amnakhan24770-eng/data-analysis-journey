```python
ruits = ["apple","mango","kiwi"]
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")
```

```python
students = ["Ali","Sara","Usman"]
for idx, name in enumerate(students, start=1):
     print(f"{idx+1}. {name}")
```

```python
names = ["Ali", "Sara", "Usman"]
scores = [90, 85, 92]

for name, sc in zip(names, scores):
    print(f"{name}: {sc}")
```

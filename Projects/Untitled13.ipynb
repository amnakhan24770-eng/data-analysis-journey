```python
print("Welcome to Amna's Simple Calculator! 👋 This calculator performs:\n •Addition \n •Subtraction \n •Multiplication \n •Division \n lets calculate!")

print("you can choose two numbers")
print("you can choose operations: ")
print("1. Addition \n2. Subtraction \n3. Multiplicaton \n4. Division")

num1= int(input("Enter number:"))
num2= int(input("Enter number:"))
choose_operation = int(input("Enter operation:"))

if choose_operation == 1:
    print(num1 + num2)
elif choose_operation == 2:
    print(num1 - num2)
elif choose_operation == 3:
    print(num1 * num2)
elif choose_operation == 4 and num2 != 0:
    print(num1 // num2)
else:
    print("invalid operation")
```

```python
import pandas as pd

sales = pd.read_csv('ecommerce_sales.csv')
```

```python
sales.shape
```

```python
sales.head()
```

```python
# new total sales column
sales['Total Sales'] = sales['Quantity'] * sales['Price']

sales
```

```python
sales.groupby('Product')['Total Sales'].sum()
```

```python
sales.groupby('Product')['Total Sales'].sum().idxmax()
```

Phone generated the highest total sales among all products in the dataset

```python
sales.groupby('Product')['Total Sales'].sum().idxmin()
```

Pen generated the lowest total sales among all products in the dataset

```python
sales.groupby('City')['Total Sales'].sum()
```

Lahore generated the highest total sales (265,000), followed by Karachi (225,000) and Islamabad (139,000). This indicates that Lahore is the strongest performing city in terms of sales revenue within this dataset.

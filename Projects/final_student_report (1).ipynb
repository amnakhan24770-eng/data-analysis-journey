```python
# new coloumn
students_record['Total Marks'] = students_record['Math'] + students_record['English'] + students_record['Science']
students_record
```

```python
students_record['Percentage'] = students_record['Total Marks']/3
students_record
```

```python
#removing total marks column

students_record = students_record.drop('total marks',axis =1)
```

```python
def grade(percentage):
    if percentage >= 80:
        return 'A'
    elif percentage >= 70:
        return 'B'
    elif percentage >= 50:
        return 'C'
    else:
        return 'Fail'

students_record['Grade'] = students_record['Percentage'].apply(grade)
students_record
```

```python
#average percentage of class

students_record.Percentage.mean()
```

```python
#finding topper
topper = students_record['Percentage'].idxmax()
```

```python
students_record.loc[topper,['Name','Percentage','City']]
```

```python
#finding lowest marks
min_marks = students_record['Percentage'].idxmin()
students_record.loc[min_marks,['Name','Percentage','City']]
```
